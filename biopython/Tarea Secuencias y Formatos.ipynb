## Ejercicio 1

```python
from Bio.Seq import Seq

# Función que encuentra orfs

def encontrarOrfs (secuencia):

    # Transcribe secuencia a proteína

    traduccion = str(secuencia.translate())
    
    orfs = []
    
    posicion = 0
    
    respaldoPosicion = 0
    
    posicion2 = 0
    
    # Recorre cada aminoacido en la secuencia
    
    for aa in traduccion:
        
        # Encuentra metionina
        
        if aa == "M":
            
            posicion2 = posicion
            
            # Recorre subsecuencua a partir de metionina
            
            for aa2 in traduccion[posicion:]:
                
                # Si encuentra otra metionina se sustituye la anterior
                
                if aa2 == "M":
                
                    posicion = posicion2
                
                # Encuentra codon de paro
                
                if aa == "M" and aa2 == "*":
                    
                    orfs.append(traduccion[posicion:posicion2+1])
                    
                    posicion = posicion2
                    
                    break
                    
                posicion2 += 1
        
        posicion +=1
    
    return(orfs)


# main()

entrada = str(input("Teclee la ruta hacia el archivo de entrada"))

archivo = open(f"{entrada}", "r").read()

print(archivo)

# Lista vacía para guardar todos los orfs
    
orfs = []

secuencia = Seq(archivo)

# Se encuentran todos los orfs de la secuencia

for marco in range(3):
    
    orfs += encontrarOrfs(secuencia[marco:])
    
    orfs += encontrarOrfs(secuencia.reverse_complement()[marco:])
    
orfMayor = ""

longitudMayor = 0

# Se compara longitud de orfs de la secuencia
    
for orf in orfs :
    
    if len(orf) > longitudMayor:
        
        longitudMayor = len(orf)
        
        orfMayor = orf
        
print(f"El orf de mayor longitud es \n {orfMayor}")
    
```

## Ejercicio 2

```python
from Bio import SeqIO

import re


# Funcion que encuentra todos los codones de una secuencia

def encontrarCodones(secuencia):
    
    codones = re.findall(r"(.{3})",str(secuencia))
    
    return (codones)



# Funcion que encuentra marcos de lectura de una secuencia

def encontrarMarcos(secuencia):

    marcos = []   
    
    secuencia = secuencia
    
    # Parte la secuencia en sus diferentes marcos forward
    
    for marco in range(3):
        
        forward = secuencia[marco:]
        
        marcos.append(str(forward))
        
    # Parte la secuencia en sus marcos reverse
        
    for marco in range(3):
    
        
        reverse = secuencia.reverse_complement()
        
        reverse = reverse[marco:]

        marcos.append(str(reverse))
        
    return(marcos)


# main()


entrada = str(input("Teclee ruta hacia archivo fasta"))

archivo = open(f"{entrada}", "r")

# Crea diccionario de archivo fasta

id_dict = SeqIO.to_dict(SeqIO.parse(archivo, 'fasta'))

numeroMarco = 0


# Recorre cada secuencia en el diccionario creado

for secuencia in id_dict:
    
    numeroMarco = 0
    
    # Obtiene id de secuencia del diccionario
    
    idSecuencia = secuencia
    
    # Obtiene secuencia del diccionario
    
    secuencia = id_dict[secuencia].seq
    
    # Obtiene marcos de lectura de la secuencia
    
    marcos = encontrarMarcos(secuencia)
    
    # Recorre la lista de marcos de lectura regresada por la funcion anterior
    
    for marco in marcos:
        
        # El contador sirve para saber qué marco de lectura es
        
        numeroMarco += 1
        
        codonesMarco = encontrarCodones(marco)
        
        # Los condicionales sirven para cambiar impresion de marcos en formato fasta
        
        if numeroMarco < 4:
        
            print(f"\n\n>{idSecuencia}_frame_{numeroMarco}")
            
            for i in codonesMarco:
                
                print(i, end = " ")
            
        if numeroMarco > 3:
            
            print(f"\n\n>{idSecuencia}_frame_minus_{numeroMarco-3}")
            
            # Se imprime cada codon
            
            for i in codonesMarco:
                
                print(i, end = " ")
```

## Ejercicio 3

```python
from Bio import SeqIO

import statistics

# Input para promedio de phred score a tomar en cuenta

phredScore = float(input("Teclea el promedio minimo a tomar en cuenta "))

entrada = str(input("Teclee ruta hacia su archivo de entrada"))

# Se crea diccionario de archivo fastq

archivo = open(f"{entrada}", "r")
    
diccionario = list(SeqIO.parse(archivo, "fastq"))
    
numeroEntradas = 0

# Recorre cada entrada del diccionario
    
for entrada in diccionario:
    
    # Calcula phred score promedio de la entrada
    
    promedio = statistics.mean(entrada.letter_annotations["phred_quality"])
    
    if  promedio < phredScore:
            
            numeroEntradas += 1
            
print(f"Hay {numeroEntradas} entradas con phred score promedio menor a {phredScore}")
    
    
```

```python
from Bio import SeqIO

entrada = str(input("Teclea ruta hacia archivo de entrada"))

archivo = open(f"{entrada}", "r")

# Se abre el archivo para leerlo

gb_record = SeqIO.read(archivo, "genbank")

archivoOutput = open("../build/outputE4Seq&Formatos.fasta", "w+")

print("Esto se escribe en archivo fasta")

# Se recorre la seccion features del archivo

for feature in gb_record.features:
    
    # Se obtiene la region CDS (donde se encuenta la traduccion) de features 
    
    if feature.type == "CDS":
        
        archivoOutput.write(f">Translation_{str(gb_record.id)}\n\n")
        
        archivoOutput.write(str(feature.qualifiers["translation"][0]))
        
        print(f">Translation_{str(gb_record.id)}\n\n")
        
        print(str(feature.qualifiers["translation"][0]))
         
```

## Ejercicio 5

```python
from Bio import PDB

import statistics

entrada = str(input("Teclea ruta hacia archivo de entrada"))


archivo = open(f"{entrada}")

# Se crea parser 

parser = PDB.PDBParser(QUIET = True)

# Se obtiene structure de archivo

struc = parser.get_structure("4hhb", archivo)

# Se asigna el atomo del residuo especificado a una variable

atomoResiduo = struc[0]["C"][20]["CA"]

# Se imprimen coordenadas del atomo

print("Coordenadas atomo: \n")

print(atomoResiduo.get_coord())

coordX = 0

coordY = 0

coordZ = 0

# Se obtienen las coordenadas de todos los atomos del residuo

for atomo in struc[0]["C"][20]:
    
    coordX += atomo.get_coord()[0]
    
    coordY += atomo.get_coord()[1]
    
    coordZ += atomo.get_coord()[2]
    
# Se obtiene el promedio de las coordenadas de los atomos
    
coordX = coordX / len(struc[0]["C"][20])

coordY = coordY / len(struc[0]["C"][20])

coordZ = coordZ / len(struc[0]["C"][20])

print("\nCoordenadas centroide:\n")

print(coordX, coordY, coordZ)
```

```python
from Bio.Blast import NCBIXML

entrada = str(input("Teclea ruta hacia archivo de entrada"))

archivo = open(entrada)

blast_records = NCBIXML.parse(archivo)

# Parsea cada componente del output de blast

for blast_record in blast_records:
    
     for alignment in blast_record.alignments:
            
        hspNumber = 0
        
        for hsp in alignment.hsps:
            
            hspNumber += 1
            
            if hsp.expect < 1e-10:
                
                print(f"\n\n>>Query name: {blast_record.query_id}")
                
                print(f"\nHit ID: {alignment.hit_id}")
                
                print("\nHit description\n")
                
                print(alignment.hit_def)
                
                print(f"\ne-Value: {hsp.expect}\n")
                
                print(f"Hsp_Number {hspNumber}\n")
```
